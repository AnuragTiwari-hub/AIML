```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing

# Load a sample dataset
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['target'] = housing.target

# Introduce some artificial missing values and outliers for demonstration
df.iloc[::100, 0] = np.nan
df.iloc[::50, 1] = np.nan
df.iloc[::200, 2] = 1000 # Artificial outlier
```

```python
# Option A: Impute missing values with the median of the column
# This is generally robust to outliers
df['MedInc'].fillna(df['MedInc'].median(), inplace=True)
df['HouseAge'].fillna(df['HouseAge'].median(), inplace=True)

# Option B: Drop rows with missing values (if the number is small)
# df.dropna(inplace=True)
```

```python
# Calculate Q1, Q3, and IQR for a specific column (e.g., 'AveRooms')
Q1 = df['AveRooms'].quantile(0.25)
Q3 = df['AveRooms'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_filtered = df[(df['AveRooms'] >= lower_bound) & (df['AveRooms'] <= upper_bound)]

print(f"Original shape: {df.shape}")
print(f"Filtered shape (after outlier removal): {df_filtered.shape}")
```
