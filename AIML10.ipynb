```python
import pandas as pd
import numpy as np

# Create a sample DataFrame (replace with your actual data loading)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Age': [25, 32, 18, np.nan, 40, 22],
    'City': ['NY', 'LA', 'Chicago', 'Houston', 'Miami', 'NY'],
    'Salary': [70000, 95000, 45000, 80000, np.nan, 60000]
}
df = pd.DataFrame(data)

# Export the DataFrame to a CSV file (for the "export" part of the problem)
df.to_csv('sample_data.csv', index=False)

# Import the data from the CSV file
df_imported = pd.read_csv('sample_data.csv')
```

```python
print(f"Number of rows: {df_imported.shape[0]}")
print(f"Number of columns: {df_imported.shape[1]}")
print("\nFirst five rows:")
print(df_imported.head(5))
print(f"\nSize (total elements): {df_imported.size}")
```

```python
print("\nNumber of missing values per column:")
print(df_imported.isnull().sum())
```

```python
# Select numerical columns
numerical_cols = df_imported.select_dtypes(include=[np.number])

print("\nStatistics for numerical columns:")
print(f"Sum:\n{numerical_cols.sum()}")
print(f"\nAverage (mean):\n{numerical_cols.mean()}")
print(f"\nMin:\n{numerical_cols.min()}")
print(f"\nMax:\n{numerical_cols.max()}")
```
