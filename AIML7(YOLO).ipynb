```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt
import numpy as np

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
  # Save the uploaded image with a consistent name
  !mv {fn} uploaded_image.jpg
  image_path = '/content/uploaded_image.jpg'
```

```python
from ultralytics import YOLO
from IPython.display import Image, display

# Load a pretrained YOLO model (e.g., yolov8n.pt)
model = YOLO('yolov8n.pt')

# Run inference on the uploaded image
results = model.predict(source=image_path, save=True, imgsz=640, conf=0.25)

# Display the results
# The results are typically saved in a 'runs/detect/predict' directory
# The first run will be in 'runs/detect/predict', subsequent runs in 'runs/detect/predict2', etc.
# The code below displays the image from the latest prediction run
import glob
from IPython.display import Image, display

# Find the directory with the saved prediction results
runs_dir = sorted(glob.glob('runs/detect/predict*'))[-1]
detected_image_path = f"{runs_dir}/uploaded_image.jpg"

print(f"Detected image saved at: {detected_image_path}")
display(Image(filename=detected_image_path))
```
