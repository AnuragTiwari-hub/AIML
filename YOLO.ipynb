```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt
import numpy as np

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
  # The uploaded file is already in /content/. We will use its original filename.
  image_path = f'/content/{fn}'

print(f"'image_path' is now set to: {image_path}")
```

```python
from ultralytics import YOLO
from IPython.display import Image, display
from pathlib import Path # Import Path for easier filename extraction

# Load a pretrained YOLO model (e.g., yolov8n.pt)
# Using 'yolov8m.pt' for potentially better accuracy (medium model)
model = YOLO('yolov8m.pt')

# Run inference on the uploaded image
# Adjusted 'conf' (confidence threshold) to a lower value for potentially more detections.
# You can experiment with this value (e.g., 0.15, 0.1)
results = model.predict(source=image_path, save=True, imgsz=640, conf=0.15)

# Display the results
# The results object contains the path to the saved image
# For a single image, results is a list with one item
if results:
    detected_image_path = results[0].path
    print(f"Detected image saved at: {detected_image_path}")
    display(Image(filename=detected_image_path))
else:
    print("No results found for object detection.")
```
